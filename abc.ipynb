### Importing libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### We are using cv2 from opencv to perform all the tasks.

### Read the image and also resize it.

```python
image=cv2.imread("D:\\cat.jpg",1)
image=cv2.resize(image, (400,400))
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title('original image',fontsize=20)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

### create new image by converting original image into gray scale image.

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow('cat',gray_img)
cv2.waitKey()
```

### Invert gray scale image.

```python
invert_img=255-gray_img
cv2.imshow("inverted_image",invert_img)
cv2.waitKey()
```

### Blur the image by using gaussianBlur funtion .

```python
blur=cv2.GaussianBlur(invert_img,(21,21),0)
```

### Invert the blurred image

```python
invert_blur=255-blur
pencil=cv2.divide(gray_img,invert_blur, scale=256.0)
cv2.imshow("sketch",pencil)
cv2.waitKey()
```

### Above step we have converted blurred image into pencil sketch.

### Here is the final output.

```python
plt.figure(figsize=(5,5))
plt.imshow(pencil,cmap="gray")
plt.title("sketch of image",fontsize=20)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```
